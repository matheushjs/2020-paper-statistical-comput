```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample100 = pd.read_csv("sample100.csv")
sample50  = pd.read_csv("sample50.csv")
sample20  = pd.read_csv("sample20.csv")
sample10  = pd.read_csv("sample10.csv")

allSamples = [sample100, sample50, sample20, sample10]
```

```python
m = sample100["data.mean"]
s = sample100["data.stddev"]

sns.set(style="whitegrid")
sns.set_palette("cubehelix")

# We do some filtering here
idx1 = np.argwhere(np.array(m > 125));
idx2 = np.argwhere(np.array(s > 100));
idx = np.unique(np.append(idx1, idx2))

print(len(idx))

for i, s in enumerate(allSamples):
    allSamples[i].drop(index=idx, inplace=True)

plt.figure(dpi=100)
m = sample100["data.mean"]
s = sample100["data.stddev"]
plt.scatter(m, s)
plt.xlabel("mean")
plt.ylabel("std dev")
plt.savefig("mean_stddev.png")
```

```python
columns = ["dist05-c{}".format(i) for i in range(1, 6)]
data = sample100[columns]
data.columns = ["min", "$c_1$", "$c_2$", "$c_3$", "$c_4$"]

plt.figure(dpi=100)
sns.set(style="whitegrid")
sns.set_palette("Set2")
sns.boxplot(data=data)
plt.title("rel. distance from 5% quantile -- n = 100")

plt.show()
```

```python
columns1 = ["dist05-c{}".format(i) for i in range(1, 6)]
columns2 = ["dist01-c{}".format(i) for i in range(1, 6)]
data1 = sample100[columns1].copy()
data2 = sample100[columns2].copy()
data1.columns = ["min", "$c_1$", "$c_2$", "$c_3$", "$c_4$"]
data2.columns = ["min", "$c_1$", "$c_2$", "$c_3$", "$c_4$"]

data1 = data1.melt()
data2 = data2.melt()

data1["quantile"] = "0.5"
data2["quantile"] = "0.1"

data = data1.append(data2)
data = data.rename(columns={"value": "relative distance"})

plt.figure(dpi=100)
sns.set_palette("cubehelix", 1)
sns.boxplot(data=data, x="variable", y="relative distance", hue="quantile", hue_order=["0.5", "0.1"])
plt.title("relative distance from 5% and 1% quantile -- n = 100")
plt.legend(loc='upper left', title="quantiles", framealpha=0.7, ncol=2)

plt.savefig("joint_boxplot_n100.png")
plt.show()
```
