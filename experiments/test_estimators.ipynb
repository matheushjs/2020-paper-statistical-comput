```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample100 = pd.read_csv("sample100.csv")
sample50  = pd.read_csv("sample50.csv")
sample20  = pd.read_csv("sample20.csv")
sample10  = pd.read_csv("sample10.csv")

allSamples = [sample100, sample50, sample20, sample10]
```

```python
for S in allSamples:
    S["dist1"] = np.abs(S["data.05"] - S["c1"]) / S["data.mean"]
    S["dist2"] = np.abs(S["data.05"] - S["c2"]) / S["data.mean"]
    S["dist3"] = np.abs(S["data.05"] - S["c3"]) / S["data.mean"]
    S["dist4"] = np.abs(S["data.05"] - S["c4"]) / S["data.mean"]
```

```python
data = sample100[["dist1", "dist2", "dist3", "dist4"]]
data.columns = ["c1", "c2", "c3", "c4"]

plt.figure(dpi=100)
sns.set(style="whitegrid")
sns.boxplot(data=data)
plt.title("rel. distance from 5% quantile -- n = 100")

plt.show()
```
